```python
from bs4 import BeautifulSoup
import requests 
import pandas as pd
```

```python
website = 'https://countrycassette.com/average-monthly-salary-by-country-2023/'
result = requests.get(website)
content = result.text

soup = BeautifulSoup(content,'lxml')
```

```python
print(soup.prettify())
```

```python
table = soup.find('table', class_='tablepress tablepress-id-150')
```

```python
data = []
for row in table.find_all('tr'):

   cols = row.find_all('td')

   # Extracting the table headers

   if len(cols) == 0:

       cols = row.find_all('th')

   cols = [ele.text.strip() for ele in cols]

   data.append([ele for ele in cols if ele])  # Get rid of empty values
```

```python
data
```

```python
# Obtain every title of columns with tag <th>
headers = data.pop(0)
df = pd.DataFrame(data, columns=('Rank', 'Country', 'Average Monthly Net Salary'))
df.index += 1
```

```python
df
```

```python
df['Average Monthly Net Salary'] = df['Average Monthly Net Salary'].str.replace('$','',regex=True)
df['Average Monthly Net Salary'] = df['Average Monthly Net Salary'].str.replace(',','',regex=True)
df
```

```python
df['Average Monthly Net Salary']
```

```python
df['Average Monthly Net Salary'] = df['Average Monthly Net Salary'].astype(float)
```

```python
Months = 12
df['Average Yearly Net Salary*'] = df['Average Monthly Net Salary'] * (Months)
countries_salary =df.round(2)
countries_salary
```

```python
del countries_salary['Average Monthly Net Salary']
```

```python
countries_salary.to_csv('average_salary.csv', encoding='utf-8')
```
