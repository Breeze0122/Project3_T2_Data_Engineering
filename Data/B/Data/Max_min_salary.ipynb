```python
# IMPORT ALL LIBRERIES
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pyarrow as py
import seaborn as se
```

```python
max_salary= pd.read_csv('/Users/anasofiagarridorivera/Desktop/Bootcamp/CHALLENGES/proyect/Project3_T2_Data_Engineering/Data/B/Data/MAX_SALARY.csv')
max_salary
```

```python
min_salary= pd.read_csv('/Users/anasofiagarridorivera/Desktop/Bootcamp/CHALLENGES/proyect/Project3_T2_Data_Engineering/Data/B/Data/MIN_SALARY.csv')
min_salary
```

```python
df_max = pd.DataFrame(data=max_salary)
df_max = df_max.rename(columns={'country':'Country', 'highest_salary':'highest_salary'})

df_max
```

```python
df_min = pd.DataFrame(data=min_salary)
df_min
```

```python
df_min2 = df_min.rename(columns={'country':'erase', 'lowest_salary':'lowest_salary'})
df_min2
```

```python
result = pd.concat([df_max, df_min2], axis=1)
result2 = result.drop('erase', axis=1)
result2
```

```python
Highest = result2.sort_values('highest_salary').reset_index().drop('index',axis=1)
Highest
```

```python
Highest['highest_salary'].dtypes
# Assuming Highest is your DataFrame
# Remove non-numeric characters and convert to float
Highest['highest_salary'] = Highest['highest_salary'].replace('[\$,]', '', regex=True).astype(float)
Highest['lowest_salary'] = Highest['lowest_salary'].replace('[\$,]', '', regex=True).astype(float)

Highest
```

```python
Highest['Median_salary'] = (Highest['highest_salary'] + Highest['lowest_salary']) / 2
Highest
```

```python
Highest.to_csv('max_min.csv', encoding='utf-8')
```
